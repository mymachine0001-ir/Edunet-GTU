```python
# Function Example

def greet():
    print("Welcome to Python")

greet()
```

```python
# If Statement Example

age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

```python
# For Loop Example

for i in range(1, 6):
    print(i)
```

```python
# Method Example (String Method)

text = "python"

print("Original:", text)
print("Upper:", text.upper())
```

```python
# List Method Example

numbers = [10, 20, 30]

print("Original:", numbers)

numbers.append(40)

print("After append:", numbers)
```
